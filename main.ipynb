```python
import pandas as pd
import tensorflow as tf
import numpy as np
```

```python
df = pd.read_csv('tmdb_5000_movies.csv')
df.head(3)
```

```python
df = df['original_title']
df
```

```python
movie_name = df.to_list()
movie_name
```

```python
tokenizer = tf.keras.preprocessing.text.Tokenizer()
tokenizer.fit_on_texts(movie_name)
seq = tokenizer.texts_to_sequences(movie_name)
```

```python
seq[:10]
```

```python
tokenizer.word_index
```

```python
X = []
y = []
total_words_dropped = 0

for i in seq:
    if len(i) > 1:
        for index in range(1, len(i)):
            X.append(i[:index])
            y.append(i[index])
    else:
        total_words_dropped += 1

print("Total Single Words Dropped are:", total_words_dropped)
```

```python
X[:10]
```

```python
y[:10]
```

```python
X = tf.keras.preprocessing.sequence.pad_sequences(X)
```

```python
X
```

```python
X.shape
```

```python
y = tf.keras.utils.to_categorical(y)
```

```python
y
y.shape
```

```python
vocab_size = len(tokenizer.word_index) + 1
vocab_size
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(vocab_size, 14),
    tf.keras.layers.LSTM(100, return_sequences=True),
    tf.keras.layers.LSTM(100),
    tf.keras.layers.Dense(100, activation='relu'),
    tf.keras.layers.Dense(vocab_size, activation='softmax'),
])
```

```python
model.summary()
```

```python
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.004),
    loss='categorical_crossentropy',
    metrics=['accuracy'])
```

```python
model.fit(X, y, epochs=150)
```

```python
model.save('nwp.h5')
```

```python
# prompt: ai modelni test qilish uchun gradio template tayyorlab ber

import gradio as gr
import pandas as pd
import tensorflow as tf
import numpy as np

# Load the model and other necessary components
# (Assuming you have saved the model as 'nwp.h5' and other relevant data)
model = tf.keras.models.load_model('nwp.h5')  # Load your trained model
df = pd.read_csv('tmdb_5000_movies.csv')
df = df['original_title']
movie_name = df.to_list()

tokenizer = tf.keras.preprocessing.text.Tokenizer()
tokenizer.fit_on_texts(movie_name)

vocab_array = np.array(list(tokenizer.word_index.keys()))


def make_prediction(text, n_words):
    for i in range(n_words):
        text_tokenize = tokenizer.texts_to_sequences([text])
        text_padded = tf.keras.preprocessing.sequence.pad_sequences(text_tokenize, maxlen=14)
        prediction = np.squeeze(np.argmax(model.predict(text_padded), axis=-1))

        # Handle cases where prediction might be out of bounds
        if prediction < len(vocab_array):
            prediction_word = str(vocab_array[prediction -1])
        else:
            prediction_word = "" #Or some default value.

        text += " " + prediction_word
    return text

# Create the Gradio interface
iface = gr.Interface(
    fn=make_prediction,
    inputs=[
        gr.Textbox(lines=1, placeholder="Enter starting text here..."),
        gr.Number(value=2, label="Number of words to predict")
    ],
    outputs=gr.Textbox(),
    title="Movie Title Prediction",
    description="Predict the next words in a movie title.",
)


iface.launch()
```
